```python
import os
import csv
import pandas as pd

filelist=os.listdir("C:/Users/Hiroka/Downloads/sta160/Dataset/csv/4_rankings")
dfs = [pd.read_csv("C:/Users/Hiroka/Downloads/sta160/Dataset/csv/4_rankings/"+f, index_col=False,  names=["week_title",
"week_year",
"week_month",
"week_day",
"rank_text",
"rank_number",
"move_positions",
"move_direction",
"player_age",
"ranking_points",
"tourneys_played",
"player_url",
"player_slug",
"player_id"],header=None) for f in filelist if f.endswith('csv')]
ranks = pd.concat(dfs, axis=0, join='outer',ignore_index=True).sort_index()
ranks
```

```python
top10=ranks[ranks["rank_number"]<=5]
top10=top10[["week_title","week_year","rank_number","player_slug","player_id"]]
top10['week_title'] = top10['week_title'].apply(lambda x: pd.to_datetime(str(x), format='%Y.%m.%d'))
top10['week_year']=top10['week_year'].astype(str).str[2]
top10
```

```python
from plotnine import *
import plotnine as p9
```

```python
(p9.ggplot(top10, aes(x='week_title',y='rank_number',group='player_slug')) +
geom_line(aes(color='player_slug'))+theme(figure_size = (11,8),legend_position="bottom",axis_text_x=element_text(rotation=45, hjust=1)))+facet_wrap('~ week_year',scales = "free")
```

```python
big4=ranks[ranks["player_slug"].isin(["novak-djokovic", "rafael-nadal","roger-federer","andy-murray"])]
big4=big4[["week_title","week_year","rank_number","player_slug","player_id"]]
big4['week_title'] = big4['week_title'].apply(lambda x: pd.to_datetime(str(x), format='%Y.%m.%d'))
big4
```
